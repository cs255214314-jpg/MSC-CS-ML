```python
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.simplefilter("ignore", DeprecationWarning)

arr = np.array([10,20,30,40,50])
arr_stand = (arr - arr.mean()) / arr.std()
print("Standardized array:", arr_stand)

from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
scaled = Scaler.fit_transform(arr.reshape(-1,1))
print("Scaled array (StandardScaler):\n", scaled)

from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
scaled_minmax = min_max.fit_transform(arr.reshape(-1,1))
print("Scaled array (MinMaxScaler):\n", scaled_minmax)

import pandas as pd
data = {
    'name':['rex','peter','sam','rita','hilda'],
    'age':[45,27,46,63,42],
    'height':[53,64,82,46,26],
    'weight':[54,53,64,76,43],
    'bp':[63,63,78,52,64]
}
frame = pd.DataFrame(data)
print(frame)

frame.to_csv('patient_data.csv', index=False)
df = pd.read_csv('patient_data.csv')
print(df)

df = df.replace(0, df.mean())
print(df)

frame2 = df[['age','bp','height','weight']]
scaled = Scaler.fit_transform(frame2)
print("Scaled features (StandardScaler):\n", scaled)
print("Mean:", scaled.mean(axis=0))
print("Std Dev:", scaled.std(axis=0))

frame3 = df[['age','bp','height','weight']]
scaled2 = min_max.fit_transform(frame3)
print("Scaled features (MinMaxScaler):\n", scaled2)
print("Mean:", scaled2.mean(axis=0))
print("Std Dev:", scaled2.std(axis=0))

from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold=(.5*(1-.9)))
useful_features1 = sel.fit_transform(scaled)
print('Useful features (StandardScaler):\n', useful_features1)

useful_features2 = sel.fit_transform(scaled2)
print('Useful features (MinMaxScaler):\n', useful_features2)
```
